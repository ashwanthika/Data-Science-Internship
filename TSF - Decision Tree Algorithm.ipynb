```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
#Changing Working Directory
import os 
os.chdir("D:\\Python")
```

```python
#Importing
ds = pd.read_csv("Iris.csv")
```

```python
ds.head()
```

```python
#Missing Value Treatment
ds.isna().sum()
```

```python
ds["Species"] = ds["Species"].astype('category')
ds["Species"] = ds["Species"].cat.codes
```

```python
ds.head()
```

```python
#Classifying Independent,Dependent variables
x = ds.iloc[:,1:5].values
y = ds.iloc[:,-1].values
```

```python
#Splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#Training
from sklearn.tree import DecisionTreeClassifier
Classifier=DecisionTreeClassifier().fit(x_train,y_train)
```

```python
#Prediction
y_pred = Classifier.predict(x_test)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
feat = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
```

```python
#Visualising
fig = plt.figure(figsize=(30,20))
tree = tree.plot_tree(Classifier, feature_names=feat, filled=True)
```
